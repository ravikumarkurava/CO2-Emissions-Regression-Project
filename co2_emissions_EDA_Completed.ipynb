### Encoding the categorical columns

```python
df['make'].nunique(), df['model'].nunique(), df['vehicle_class'].nunique()
```

```python
df['cylinders'].nunique(), df['transmission'].nunique(), df['fuel_type'].nunique()
```

#### By observing the above unique number of values for each categorical columns we have decided that,
#### One_Hot_Encoding for cylinders, transmission and fuel_type.
#### Frequency_Encoding for make, model, vehicle_class.

```python
df = pd.get_dummies(df, columns=['cylinders', 'transmission', 'fuel_type'])
```

```python
make_freq = df['make'].value_counts(normalize=True)
df['make_freq'] = df['make'].map(make_freq)
```

```python
model_freq = df['model'].value_counts(normalize=True)
df['model_freq'] = df['model'].map(model_freq)
```

```python
vehicle_class_freq = df['vehicle_class'].value_counts(normalize=True)
df['vehicle_class_freq'] = df['vehicle_class'].map(vehicle_class_freq)
```

```python
df.drop(columns=['make', 'model', 'vehicle_class'], inplace=True)
```

```python
df.head(2)
```

```python
df.info()
```

```python
target = df.pop('co2_emissions')
df['co2_emissions'] = target
```

```python
num_cols = ['engine_size', 'fuel_consumption_city', 'fuel_consumption_hwy', 'fuel_consumption_comb(l/100km)', 'fuel_consumption_comb(mpg)']
print(num_cols)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```
