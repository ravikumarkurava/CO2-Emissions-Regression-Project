```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('co2_emissions_EDA_Completed.csv')
```

```python
df.head(2)
```

```python
df.info()
```

#### split data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:,:26]
y = df.iloc[:,26]
```

```python
X.info()
```

```python
X.head(2)
```

```python
y.info()
```

```python
y.head(2)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

### Create the model

#### Find the right value of k

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
train_mse = []
test_mse = []

for i in range(1,20) : 
    knn_reg = KNeighborsRegressor(n_neighbors=i)
    knn_reg.fit(X_train, y_train)
    tr_pred = knn_reg.predict(X_train)
    ts_pred = knn_reg.predict(X_test)
    tr_mse = mean_squared_error(y_train, tr_pred)
    ts_mse = mean_squared_error(y_test, ts_pred)
    train_mse.append(tr_mse)
    test_mse.append(ts_mse)
```

```python
plt.figure(figsize=(12,5))
plt.title('Finding right value of k')
plt.xlabel('Num of neighbors')
plt.ylabel('mean squared error')
plt.plot(train_mse, label='train')
plt.plot(test_mse, label='test')
plt.xticks(range(0,20))
plt.legend()
plt.show()
```

#### Considerig the above graph, we will select k=5

```python
model = KNeighborsRegressor(n_neighbors=5)
```

```python
model.fit(X_train, y_train)
```

#### Predict for training Data and get the metrics

```python
yhat_train = model.predict(X_train)
```

```python
train_mse = mean_squared_error(y_train,yhat_train)
train_rmse = np.sqrt(train_mse)
train_r2 = r2_score(y_train,yhat_train)
train_mse, train_rmse, train_r2
```

```python
y_train.mean()
```

#### Predict for testing Data and get the metrics

```python
yhat_test = model.predict(X_test)
```

```python
test_mse = mean_squared_error(y_test,yhat_test)
test_rmse = np.sqrt(test_mse)
test_r2 = r2_score(y_test,yhat_test)
test_mse, test_rmse, test_r2
```
